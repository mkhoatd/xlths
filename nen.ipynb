```python
# read file by librosa
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from __future__ import annotations

# read audio.mp4a
y: np.ndarray  # âm thanh
sr: int  # tần số lấy mẫu
(y, sr)=librosa.load('b.m4a')
M=10
new_y=np.array([y[i] for i in range(0, len(y), M)]) #new_y[2]=y[2M]
y[0:20], new_y[0:20]
```

```python
len(y)/sr
```

```python
import sounddevice as sd
```

```python
sd.play(y, sr)
```

```python
sd.play(new_y, sr//M)
```

```python
fig, (ax0, ax1)=plt.subplots(nrows=1,
                       ncols=2,
                       figsize=(14, 5))
librosa.display.waveshow(y, sr=sr, ax=ax0)
librosa.display.waveshow(new_y, sr=sr//M, ax=ax1)
plt.show()
```
