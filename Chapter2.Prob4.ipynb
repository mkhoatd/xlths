```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_start = 0
t_end = 2
Fs = 8000
T0 = 1/Fs
F0 = 1000
A = 2
ph = 0
```

```python
Fs1 = 3* F0
Fs2 = 1.5*F0
```

```python
T1 = 1/Fs1
T2 = 1/Fs2
```

```python
t0 = np.arange(t_start, t_end, T0)
t1 = np.arange(t_start, t_end, T1)
t2 = np.arange(t_start, t_end, T2)
```

```python
x0 = 2*np.cos(2*np.pi*F0*t0 + ph)
x1 = 2*np.cos(2*np.pi*F0*t1 + ph)
x2 = 2*np.cos(2*np.pi*F0*t2 + ph)
```

```python
fig, ((ax1, ax2)) = plt.subplots(nrows=1,
                                 ncols=2,
                                 figsize = (12,7)) ;
# plot
ax1.plot(t1[0:100],x1[0:100])

# plot
ax2.plot(t2[0:50],x2[0:50])
```
